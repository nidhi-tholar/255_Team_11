```python
import pandas as pd
Data = pd.read_csv('C:/Users/sasha/Documents/#Classes/CMPE-255 Carlos Rojas/Project_Network/output.csv')

Data['Opening'].unique()

Moves=Data['Moves']
```

```python
from difflib import SequenceMatcher

matchingGame=[]

for i in range(0,20):
  dum=[]
  for j in range(0,25):
    dum.append(SequenceMatcher(None,Moves[i],Moves[j]).ratio())
  
  matchingGame.append(dum)
```

```python
MovRel = pd.DataFrame(matchingGame)
MovRel
```

```python
CheckGame = 0
```

```python
Top2=pd.DataFrame(MovRel.loc[CheckGame,:].nlargest(2))
```

```python
Top2
```

```python
if 1.0 in list(Top2[0]):
    Indx = Top2[Top2[0]!=1.0].index[0]
else:
    Indx = Top2[Top2[0]==Top2[0].max()].index[0]
```

```python
Indx
```

```python
print('Game Considered :\n',Moves[CheckGame],'\n\n\nGame With higher similarity:\n',Moves[Indx])
```
